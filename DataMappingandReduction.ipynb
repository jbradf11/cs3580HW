```python
import pandas as pd
import matplotlib.pyplot as plt
dft = pd.read_csv('/mnt/c/Users/jrbra/OneDrive/Documents/Skool/csis3580/acs2015_census_tract_data.csv', encoding='latin-1')
dfc = pd.read_csv('/mnt/c/Users/jrbra/OneDrive/Documents/Skool/csis3580/acs2015_county_data.csv', encoding='latin-1')
```

```python
dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').Black.agg(['mean']).sort_index()
```

```python
#Change Columns from % to Total
dfc.Hispanic = dfc.TotalPop * (dfc.Hispanic /100)
dfc.White = dfc.TotalPop * (dfc.White /100)
dfc.Black = dfc.TotalPop * (dfc.Black /100)
dfc.Native = dfc.TotalPop * (dfc.Native /100)
dfc.Asian = dfc.TotalPop * (dfc.Asian /100)
dfc.Pacific = dfc.TotalPop * (dfc.Pacific /100)
```

```python
# aggregate state totals
BlackPBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').Black.agg(['sum']).sort_index()
HispanicPBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').Hispanic.agg(['sum']).sort_index()
WhitePBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').White.agg(['sum']).sort_index()
NativePBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').Native.agg(['sum']).sort_index()
AsianPBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').Asian.agg(['sum']).sort_index()
PacificPBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').Pacific.agg(['sum']).sort_index()
```

```python
# Create dataframe that is just State and TotalPop
PBS = dfc[(dfc.State != 'Puerto Rico') & (dfc.State != 'District of Columbia')].groupby('State').TotalPop.agg(['sum']).sort_index()
PBS
```

```python
# Merge Datasets and assign good names to columns
PBSDiversity = pd.merge(PBS, BlackPBS, on='State')
PBSDiversity = pd.merge(PBSDiversity, HispanicPBS, on='State')
PBSDiversity = PBSDiversity.rename(columns={'sum_x':'TotalPop','sum_y':'BlackPop','sum':'HispanicPop'})
PBSDiversity = pd.merge(PBSDiversity, WhitePBS, on='State')
PBSDiversity = pd.merge(PBSDiversity, NativePBS, on='State')
PBSDiversity = pd.merge(PBSDiversity, AsianPBS, on='State')
PBSDiversity = PBSDiversity.rename(columns={'sum_x':'WhitePop','sum_y':'NativePop','sum':'AsianPop'})
PBSDiversity = pd.merge(PBSDiversity, PacificPBS, on='State')
PBSDiversity = PBSDiversity.rename(columns={'sum':'PacificPop'})
```

```python
# PBSDiversity
```

```python
# Create Columns that turn population totals into percents
PBSDiversity['BlackPct'] = PBSDiversity.BlackPop / PBSDiversity.TotalPop
PBSDiversity['HispanicPct'] = PBSDiversity.HispanicPop / PBSDiversity.TotalPop
PBSDiversity['WhitePct'] = PBSDiversity.WhitePop / PBSDiversity.TotalPop
PBSDiversity['NativePct'] = PBSDiversity.NativePop / PBSDiversity.TotalPop
PBSDiversity['AsianPct'] = PBSDiversity.AsianPop / PBSDiversity.TotalPop
PBSDiversity['PacificPct'] = PBSDiversity.PacificPop / PBSDiversity.TotalPop
```
